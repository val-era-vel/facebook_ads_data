```python
import pandas as pd
data = pd.read_csv('C:\\Users\\lera\\Downloads\\facebook_ads_data (2.0).csv')
data.head()
```

```python
import matplotlib.pyplot as plt
data['ad_date'] = pd.to_datetime(data['ad_date'])
data_2021 = data[data['ad_date'].dt.year == 2021]
daily_spend = data_2021.groupby(data_2021['ad_date'].dt.date)['total_spend'].sum()
plt.figure(figsize=(14, 7))
plt.plot(daily_spend.index, daily_spend.values, marker='o', linestyle='-')
plt.title('Щоденна сума витрат на рекламу в 2021 році')
plt.xlabel('Дата')
plt.ylabel('Сума витрат')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
data['ad_date'] = pd.to_datetime(data['ad_date'], errors='coerce')

data['romi'] = data['total_value'] / data['total_spend']


daily_romi = data.groupby(data['ad_date'].dt.date)['romi'].mean()


plt.figure(figsize=(14, 7))
plt.plot(daily_romi.index, daily_romi.values, marker='o', linestyle='-', color='orange')
plt.title('Щоденний ROMI в 2021 році')
plt.xlabel('Дата')
plt.ylabel('ROMI')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Графік з щоденним ROMI в 2021 році
rolling_spend = data.total_spend.rolling(window=7).mean()
rolling_romi = data.romi.rolling(window=7).mean()


plt.figure(figsize=(14, 7))
plt.plot(daily_spend.index, daily_spend.values, label='Щоденні витрати', color='green')
plt.plot(rolling_spend.index, rolling_spend.values, label='Рухоме середнє витрат', color='red')
plt.title('Рухоме середнє щоденних витрат')
plt.xlabel('Дата')
plt.ylabel('Сума витрат')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()


plt.figure(figsize=(14, 7))
plt.plot(daily_romi.index, daily_romi.values, label='Щоденний ROMI', color='orange')
plt.plot(rolling_romi.index, rolling_romi.values, label='Рухоме середнє ROMI', color='green')
plt.title('Рухоме середнє щоденного ROMI')
plt.xlabel('Дата')
plt.ylabel('ROMI')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
# групування даних за назвою кампанії
campaign_spend = data.groupby('campaign_name')['total_spend'].sum()

plt.figure(figsize=(14, 7))
campaign_spend.plot(kind='bar', color='skyblue')
plt.title('Загальна сума витрат на рекламу в кожній кампанії')
plt.xlabel('Назва кампанії')
plt.ylabel('Сума витрат')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Розрахунок ROMI для кампаній
campaign_romi = data.groupby('campaign_name')['romi'].mean()
plt.figure(figsize=(14, 7))
campaign_romi.plot(kind='bar', color='lightgreen')
plt.title('Загальний ROMI в кожній кампанії')
plt.xlabel('Назва кампанії')
plt.ylabel('ROMI')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Box plot для ROMI за кампанією
plt.figure(figsize=(14, 7))
data.boxplot(column='romi', by='campaign_name', grid=False)
plt.title('Розкид щоденного ROMI в кожній кампанії')
plt.suptitle('')
plt.xlabel('Назва кампанії')
plt.ylabel('ROMI')
plt.xticks(rotation=45)
plt.show()
```

```python
# Гістограма ROMI
plt.figure(figsize=(14, 7))
plt.hist(data['romi'], bins=30, color='purple', alpha=0.7)
plt.title('Гістограма розподілу ROMI')
plt.xlabel('ROMI')
plt.ylabel('Частота')
plt.grid()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
numeric_data = data.select_dtypes(include=['number'])

correlation_matrix = numeric_data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Теплова карта кореляції')
plt.show()

if 'total_value' in correlation_matrix.columns:
    total_value_correlations = correlation_matrix['total_value']

    highest_correlation = total_value_correlations[total_value_correlations != 1].idxmax()
    lowest_correlation = total_value_correlations[total_value_correlations != 1].idxmin()

    print(f'Найвища кореляція з total_value: {highest_correlation} ({total_value_correlations[highest_correlation]:.2f})')
    print(f'Найнижча кореляція з total_value: {lowest_correlation} ({total_value_correlations[lowest_correlation]:.2f})')
else:
    print("Стовпця 'total_value' немає в числових даних.")
```

```python
# Точкова діаграма
sns.lmplot(x='total_spend', y='total_value', data=data)
plt.title('Звʼязок між витратами та значеннями')
plt.xlabel('Загальні витрати')
plt.ylabel('Загальна вартість')
plt.show()
```
